```python
graph = {
    'A' : ['B', 'C'],
    'B' : ['D', 'E'],
    'C' : ['F'],
    'D' : [],
    'E' : ['F'],
    'F' : ['H'],
    'G' : ['H'],
    'H' : ['G']
}

visied = []
queue = []

def bfs(visited, graph, node):
    visied.append(node)
    queue.append(node)

    while queue:
        s = queue.pop(0)
        print(s, end = " ")

        for neighbour in graph[s]:
            if neighbour not in visied:
                visied.append(neighbour)
                queue.append(neighbour)

print("hasil penelusuran graf menggunakan bfs: ")
bfs(visied, graph, 'A')
```
