```python
# BFS / queue / first in first out

graph = {

    'A' : ['B'], 
    'B' : ['F', 'G'], 
    'F' : ['E'], 
    'E' : [], 
    'G' : ['C'], 
    'C' : ['D'], 
    'D' : ['H'], 
    'H' : [], 

}

visied = []
queue = []

def bfs(visited, graph, node):
    visied.append(node)
    queue.append(node)

    while queue:
        s = queue.pop(0)
        print(s, end = " ")

        for neighbour in graph[s]:
            if neighbour not in visied:
                visied.append(neighbour)
                queue.append(neighbour)

print("hasil penelusuran graf menggunakan bfs: ")
bfs(visied, graph, 'A')
```
